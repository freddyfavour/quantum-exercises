# Creating and Printing Two Random 5-Dimensional Vectors

```python
import numpy as np
```

```python
np.random.seed(42)
u = np.random.randint(-15, 16, 5)
v = np.random.randint(-15, 16, 5)
```

```python
print("Vector u (5D):")
print(u)
print("\nVector v (5D):")
print(v)
```

# Finding the Vector $(3\mathbf{u} - 2\mathbf{v})$

```python
three_u = 3 * u
two_v = 2 * v
```

```python
result_vector = three_u - two_v
```

```python
print("The vectors used:")
print(f"u: {u}")
print(f"v: {v}")
print("\nResult of (3u - 2v):")
print(result_vector)
```

# Verifying the Norm Property

```python
import numpy as np

# 1. Define the 2-dimensional vector u
u = np.array([-2, -4])

# 2. Calculate the norm of u (||u||)
norm_u = np.linalg.norm(u)

# 3. Calculate 4 * ||u|| (Right-Hand Side, RHS)
rhs = 4 * norm_u

# 4. Calculate the vector 4u
vector_four_u = 4 * u

# 5. Calculate the norm of 4u (||4u||) (Left-Hand Side, LHS)
norm_four_u = np.linalg.norm(vector_four_u)
lhs = norm_four_u
```

```python
# 6. Verify the equality
print(f"Vector u: {u}")
print(f"Vector 4u: {vector_four_u}")
print("-" * 30)

print(f"||u||: {norm_u:.4f}")
print(f"4 * ||u|| (RHS): {rhs:.4f}")
print(f"||4u|| (LHS): {lhs:.4f}")
```

```python
# Verification check
if np.isclose(lhs, rhs):
    print("\nVerification successful: ||4u|| = 4 * ||u|| is TRUE.")
else:
    print("\nVerification failed.")
```
