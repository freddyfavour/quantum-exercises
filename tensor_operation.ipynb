```python
import numpy as np
```

```python
# Define the vectors u and v
u = np.array([-2, -1, 0, 1])
v = np.array([1, 2, 3])
```

```python
# Calculate Tensor Product u (tensor) v
# This corresponds to the Kronecker product
u_tensor_v = np.kron(u, v)
```

```python
# Calculate Tensor Product v (tensor) u
v_tensor_u = np.kron(v, u)
```

```python
print("Vector u:", u)
print("Vector v:", v)
print("-" * 20)
print("u ⊗ v Result (Length 12):")
print(u_tensor_v)
print("-" * 20)
print("v ⊗ u Result (Length 12):")
print(v_tensor_u)
```
