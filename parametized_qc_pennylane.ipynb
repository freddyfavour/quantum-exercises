```python
import pennylane as qml
from pennylane import numpy as np
```

###  1. Create a device with 1 qubit

```python
dev = qml.device("default.qubit", wires=1)
```

### 2. Define the QNode (Quantum Circuit)

```python
@qml.qnode(dev)
def parameterized_circuit(theta):
    # Apply Ry(theta) and Rz(theta)
    # Hints suggest using qml.Ry and qml.Rz 
    qml.RY(theta, wires=0)
    qml.RZ(theta, wires=0)
    
    # Return the statevector [cite: 44]
    return qml.state()
```

### 3. Test with theta = pi/4 [cite: 41]

```python
theta_val = np.pi / 4
state_vector = parameterized_circuit(theta_val)

print(f"Input Theta: {theta_val}")
print("Resulting State Vector:")
print(state_vector)
```
